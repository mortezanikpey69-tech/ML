```python
import matplotlib.pyplot as plt
import seaborn as sns;sns.set()
import numpy as np
from sklearn.cluster import KMeans
```

```python
from sklearn.datasets import make_blobs
X,y_true=make_blobs(n_samples=400,centers=4,cluster_std=0.50,random_state=0)
#make dataset
```

```python
y_true[0]
#location of clustring
```

```python
plt.scatter(X[:,0],X[:,1],s=20)
plt.show()
```

```python
kmeans=KMeans(n_clusters=4)
```

```python
kmeans.fit(X)
```

```python
out_robot = kmeans.predict(X)
```

```python
#data in reality
plt.scatter(X[:,0],X[:,1],c=y_true,s=10,cmap='cool')
centers=kmeans.cluster_centers_
plt.scatter(centers[:,0],centers[:,1],c='blue',s=100,alpha=0.9)
plt.show()
```

```python
#data with robot perdict
plt.scatter(X[:,0],X[:,1],c=out_robot,s=10,cmap='cool')
centers=kmeans.cluster_centers_
plt.scatter(centers[:,0],centers[:,1],c='blue',s=100,alpha=0.9)
plt.show()
```
